```python
##From: https://pythonprogramming.net/3d-bar-charts-python-matplotlib/


%matplotlib notebook

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax1 = fig.add_subplot(111, projection='3d')

xpos = [1,25,40,40]
ypos = [1, 15,27,50]
num_elements = len(xpos)
zpos = [0,0,0,0]
dx = [(12.045439*1.5),(8.372574*1.5),(7.113368*1.5),(12.045439*1.5)]
dy = [(12.045439*1.5),(8.372574*1.5),(7.113368*1.5),(12.045439*1.5)]
dz = [150,43,60,141]

names = ['Potential', 'Shallowing', 'Deepening', 'Contemporary']
ax1.bar3d(xpos, ypos, zpos, dx, dy, dz, color= ['honeydew', 'lightcoral','cornflowerblue','lightyellow'])
ax1.set_ylim(0,70)
ax1.set_zlim(0,150)
ax1.set_zlabel('Depth (cm)')
ax1.set_xlabel('')
#ax1.set_xlabel('X * Y = Area of rooting, km2 x 6.667^3')
sea = plt.Rectangle((0, 0), 1, 1, fc="honeydew")
red_proxy = plt.Rectangle((0, 0), 1, 1, fc="lightcoral")
blue_proxy = plt.Rectangle((0, 0), 1, 1, fc="cornflowerblue")
green_proxy = plt.Rectangle((0, 0), 1, 1, fc="lightyellow")
ax1.legend([sea, red_proxy, blue_proxy, green_proxy], ['Potential', 'Shallowing','Deepening','Contemporary'], loc = 1, bbox_to_anchor=(1.3, 1.1))
plt.figtext(0.5,0.04, "X * Y = Area of rooting, km2 x 4.445^3", ha="center", va="center", fontsize=10)
plt.show()
```
