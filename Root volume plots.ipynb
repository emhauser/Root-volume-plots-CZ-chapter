```python
##From: https://pythonprogramming.net/3d-bar-charts-python-matplotlib/


#%matplotlib notebook

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax1 = fig.add_subplot(111, projection='3d')

xpos = [1,23,43,58]
ypos = [20,20,20,20]
num_elements = len(xpos)
zpos = [0,0,0,0]
dx = [12.045439,8.372574,7.113368,12.045439]
dy = [12.045439,8.372574,7.113368,12.045439]
dz = [150,43,60,141]

names = ['Potential', 'Shallowing', 'Deepening', 'Contemporary']
ax1.bar3d(xpos, ypos, zpos, dx, dy, dz, color= ['#00ceaa', 'red','blue','green'])
ax1.set_ylim(0,60)
ax1.set_zlim(0,150)
ax1.set_zlabel('Depth (cm)')
ax1.set_xlabel('')
ax1.set_xlabel('X * Y = Area of rooting, km2 x 10^3')
sea = plt.Rectangle((0, 0), 1, 1, fc="#00ceaa")
red_proxy = plt.Rectangle((0, 0), 1, 1, fc="red")
blue_proxy = plt.Rectangle((0, 0), 1, 1, fc="blue")
green_proxy = plt.Rectangle((0, 0), 1, 1, fc="green")
ax1.legend([sea, red_proxy, blue_proxy, green_proxy], ['Potential', 'Shallowing','Deepening','Contemporary'], loc = 4)
plt.show()
```
